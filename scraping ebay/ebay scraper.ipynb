#### BASIC EBAY SCRAPER

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ebay.com/sch/garlandcomputer/m.html"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")

for item in soup.find_all("div", class_="s-item__info clearfix"):
    title = item.find("div", class_="s-item__title").text
    price = item.find("span", class_="s-item__price").text
    print(f"Title: {title}\nPrice: {price}\n")
```

####SCRAPER USING IP ROTATION IN A PROXY LIST

```python
import requests
from bs4 import BeautifulSoup
from itertools import cycle

# List of proxy IP addresses
proxies = [
    'https://168.205.102.26:8080',
    'http://213.217.30.69:3128',
    'http://186.96.50.113:999',
    'http://195.114.209.50:80',
    'http://159.203.61.169:3128',
]

#defining our url and headers
url = "https://www.ebay.com/sch/garlandcomputer/m.html"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

#looping through each proxy of our list
for proxy in proxies:
    print("Using proxy:", proxy)
    
    try:
        # Make the request using the current proxy
        response = requests.get(url, headers=headers, proxies={"http": proxy, "https": proxy})
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "lxml")

            for item in soup.find_all("div", class_="s-item__info"):
                title_element = item.find("div", class_="s-item__title")
                if title_element:
                    title = title_element.text.strip()
                else:
                    title = "Title not found"
                
                price_element = item.find("span", class_="s-item__price")
                if price_element:
                    price = price_element.text.strip()
                else:
                    price = "Price not found"

                print(f"Title: {title}\nPrice: {price}\n")
            
            break  # Exit the loop if successful
        else:
            print("Failed to retrieve data. Status code:", response.status_code)
    except Exception as e:
        print("Error:", e)
        # Move to the next proxy if an error occurs
        continue
```
